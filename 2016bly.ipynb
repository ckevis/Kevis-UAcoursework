SN2016bly was discovered on 04/09/2016 when it hit peak magnitude. Its apparent magnitude at peak was m=18.17, which is an absolute magnitude of M=-16.51. It is located at an RA/Dec of 17:22:48.9/+14:00:59.89 near the center of an anonymous galaxy . We have 4 spectra taken ~+60 and +90 days from peak. These spectra were corrected using the IRAF function dopcor for an extinction value E(B-V) of 0.1256 +/- .003 (Schaifley et al 2011) and the IRAF function deredden for a redshift value of z=0.0194.

```python
#plotting the spectra of 2016bly
#there are 4 spectra taken
#peak was on 04/09/2016
#we have spectra taken on
    #06/07/2016 +59
    #06/13/2016 +65
    #07/13/2016 +95
    #07/14/2016 +96
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u

#graph all of the spectra
first = np.genfromtxt('0607-Bok.txt')
second = np.genfromtxt('0613-Bok')
third = np.genfromtxt('0713-Bok')
fourth = np.genfromtxt('0714-Bok')

fig = plt.figure(figsize=(10,10))

#the y axis is in arbitrary units of flux
#we will need to separate them with a constant in order to get a visual
#otherwise the spectra will all be on top of each other
#we also need to normalize the flux
x67 = first[:,0]
y67 = first[:,1]/4e-16

x613 = second[:,0]
y613 = second[:,1]/1e-15 + 2

x713 = third[:,0]
y713 = third[:,1]/2e-16 

x714 = fourth[:,0]
y714 = fourth[:,1]/2e-16

#the spectra will look flat if the y axis is not in log
#plt.yscale('log')

#now we have to label
plt.title('The Spectra of SN2016bly')
plt.ylabel('Flux [arbitrary units]')
plt.xlabel('Wavelength [Angstroms]')

plt.plot(x67,y67,'r-',label='+59')
plt.plot(x613,y613,'y-',label='+65')
plt.plot(x713,y713,'g-',label='+95')
#plt.plot(x714,y714,'b-',label='+96')

plt.legend(loc='lower left')

plt.show()
```

The P-Cygni profile for H-alpha appears to be shrinking in EW as time passes despite the EW of the H-alpha emission line remaining relatively the same. We can see that as time passes most of the absorption lines begin to fade which is expected. For +59, there are very defined absorption lines at ~5000A. These correspond to H-Beta [4860], FeII [5200], and OIII [5007].
